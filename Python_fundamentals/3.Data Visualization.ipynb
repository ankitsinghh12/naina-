# Python Matplotlib

**Matplotlib is a python library used for data visualization. You can create bar-plots, scatter-plots, histograms and a lot more with matplotlib.**

```python
#line plot

import numpy as np
from matplotlib import pyplot as plt

x=np.arange(1,11)
y=2*x
print (x)
print (y)
```

```python
plt.plot(x,y)
plt.title("Line Plot")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.plot(x,y,color='g',linestyle=':',linewidth=2)          #changing line aesthetics
plt.show()
```

```python
#adding two lines in the same plot

x=np.arange(1,11)
y1=2*x
y2=3*x

plt.plot(x,y1,color='g',linestyle=':',linewidth=2)
plt.plot(x,y2,color='r',linestyle='-',linewidth=3)
plt.title("Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show
```

```python
#adding sub-plots

x=np.arange(1,11)
y1=2*x
y2=3*x

plt.subplot(1,2,1)
plt.plot(x,y1,color='g',linestyle=':',linewidth=2)

plt.subplot(1,2,2)
plt.plot(x,y2,color='r',linestyle=':',linewidth=2)

plt.show()
```

# Bar Plot
**when we study categorical value distribution, we use bar plot**

```python
student = {"Bob":34,"Matt":35,"Sam":74}

names = list(student.keys())
values = list(student.values())

plt.bar(names,values)
plt.show()
```

```python
#Adding Title and Labels

plt.bar(names,values)
plt.title("Bar plot")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

```python
#Horizontal Bar Plot

plt.barh(names,values,color='g')
plt.title("Bar Plot")
plt.xlabel("Names")
plt.ylabel;("Marks")
plt.grid(True)
plt.show()
```

# Scatter Plot

```python
x=[10,20,30,40,50,60,70,80,90]
a=[8,5,3,7,2,7,5,9,4]

plt.scatter(x,a)
plt.show()
```

```python
#Changing Marks Aesthetics

x=[10,20,30,40,50,60,70,80,90]
a=[8,5,3,7,2,7,5,9,4]

plt.scatter(x,a,marker="*",c="g",s=100)
plt.show()
```

```python
#Adding two markers in the same plot

x=[10,20,30,40,50,60,70,80,90]
a=[8,5,3,7,2,7,5,9,4]
b=[3,5,6,2,4,7,9,0,3]

plt.scatter(x,a,marker="*",c="g",s=100)
plt.scatter(x,b,marker=".",c="r",s=200)
plt.show()
```

```python
#adding sub-plots

x=[10,20,30,40,50,60,70,80,90]
a=[8,5,3,7,2,7,5,9,4]
b=[3,5,6,2,4,7,9,0,3]

plt.subplot(1,2,1)
plt.scatter(x,a,marker="*",c="g",s=100)

plt.subplot(1,2,2)
plt.scatter(x,b,marker=".",c="r",s=200)
plt.show()
```

# Box-Plot
**A box and whisker plot-also called a box plot displays the five number summary of a set of data. the five number summary is the minimum, first quartile, median, third quartile and maximum. in a box plot, we draw from the first quartile to third quartile. a vertical line goes through the box at the median. the whiskers go from each quartile to the minimum or maximum.**

```python
#creating data

one = [1,2,3,4,5,6,7,8,9]
two = [1,2,3,4,5,4,3,2,1]
three = [6,7,8,9,8,7,6,5,4]

data = list([one,two,three])
plt.boxplot(data)
plt.show()
```

# Violin Plot

```python
one = [1,2,3,4,5,6,7,8,9]
two = [1,2,3,4,5,4,3,2,1]
three = [6,7,8,9,8,7,6,5,4]

data = list([one,two,three])
plt.violinplot(data,showmedians=True)
plt.show()
```

# Seaborn Line Plot
**Seaborn is an open-source Python library built on top of matplotlib. It is used for data visualization and exploratory data analysis. Seaborn works easily with dataframes and the Pandas library. The graphs created can also be customized easily.**

```python
import seaborn as sns 
from matplotlib import pyplot as plt

fmri = sns.load_dataset("fmri")                  #fmri is a inbuilt dataset in Seaborn
fmri.head()
```

```python
sns.lineplot(x="timepoint",y="signal",data=fmri)
plt.show()
```

```python
#Grouping two different events with 'hue'

#In seaborn, the hue parameter represents which column in the data frame, you want to use for color encoding.

sns.lineplot(x="timepoint",y="signal", data=fmri, hue="event")
plt.show
```

```python
#Adding styles

sns.lineplot(x="timepoint",y="signal", data=fmri, hue="event",style="event")
plt.show
```

```python
#Adding Markers

#Adding markers to a line plot can be a useful way to distinguish multiple lines or to highlight particular data points. ...

sns.lineplot(x="timepoint",y="signal", data=fmri, hue="event",style="event", markers=True)
plt.show
```

```python
#Bar Plot

sns.set(style="whitegrid")                             #for setting the white grid behind the bar plot 
sns.barplot(x="timepoint",y="signal",data=fmri)
plt.show()
```

```python
fmri.drop('subject',axis=1)            #dropping the first column
```

```python
#representing 'region' column with different colours with the help of 'hue' command

sns.barplot(x="timepoint",y="signal", data=fmri, hue="region")
plt.show
```

# Paletter

**We will classify the different ways for using color_palette() types **

**Qualitative** - A qualitative palette is used when the variable is categorical in nature, the color assigned to each group 
                  need to be distinct. Each possible value of the variable is assigned one color from a qualitative palette
                  within a plot as shown in figure.
                  
**Sequential**  - In sequential palettes color moved sequentially from a lighter to a darker. When the variable assigned to be 
                  colored is numeric or has inherently ordered values, then it can be depicted with a sequential palette as 
                  shown in figure.
                  
**Diverging**   - When we work on mixed value like +ve and -ve(low and high values) then diverging palette is the best suit for 
                  visualization.

```python
sns.barplot(x="timepoint",y="signal", data=fmri, palette='Blues_d')                 #blue palette
plt.show
```

```python
sns.barplot(x="timepoint",y="signal", data=fmri, palette='rocket')                 #rocket palette
plt.show
```

```python
sns.barplot(x="timepoint",y="signal", data=fmri, palette='vlag')                 #vlag palette
plt.show
```

```python
#if we dont want palette or default colors and we want fixed color for both the bars

sns.barplot(x="timepoint",y="signal", data=fmri, color="orange")                 #blue palette
plt.show
```
